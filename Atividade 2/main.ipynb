# Importação de bibliotecas

```python
import numpy as np
from metrics import accuracy_score
from models import GaussianNaiveBayes as GNBayes, QuadraticDiscrimimantAnalisys as QDA, LogisticRegression as LR
from sklearn.model_selection import train_test_split
from plots import Plots
```

# Leitura dos dados

```python
data = np.loadtxt("./ex2data1.txt", delimiter=",")
X = data[:, :-1]
y = data[:, -1]
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, shuffle = True, stratify=y, random_state=18)
```

# Apresentação dos resultados

```python
def apply_classifier(classifier, clf_name):
    print('________________' + clf_name +'________________')
    classifier.fit(X_train,y_train)
    y_pred = classifier.predict(X_test)
    print('Acurácia: {}'.format(accuracy_score(y_test, y_pred)))
    Plots().plot_confusion_matrix_(X_test, y_test, classifier)
    print('_____________Dados com fronteira de separação_____________')
    Plots().plot_boundaries(X_test, y_test, classifier)
```

```python
model1 = LR(epochs=300, learning_rate=0.00002)
model1.classes_ = ['X1', 'X2']
model2 = GNBayes()
model3 = QDA()
```

```python
apply_classifier(model1, "Regressão logística")
```

```python
apply_classifier(model2, "Naive bayes Gaussiano")
```

```python
apply_classifier(model3, "Analise de discriminante quadratico")
```

# Relatório

## O dado parece se linearmente separável ou não?
Sim, os dados são linearmente separáveis.

## Quais dos métodos produziu fronteiras lineares?
A regressão logística e o naive bayes gaussiano. 

## Qual teve melhor acurácia?
O naive bayes Gaussiano obteve o melhor valor de acurácia, com o discriminante quadrático gaussiano obtendo valores próximos.
