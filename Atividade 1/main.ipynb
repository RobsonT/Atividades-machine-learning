## Importação das bibliotecas

```python
from regression import (RLA, RLGD, MLR, MLRGD, MLRSGD, Polynomial_regression as PR,
Regularized_multiple_linear_regression as RMLR)
from metrics import Metrics
import numpy as np
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
```

## Leitura do Boston House Price Dataset

```python
data = np.loadtxt("./data/housing.data", delimiter=",")
X = data[:,-2]
y = data[:, -1]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, shuffle=True, test_size=0.2)
```

## Aplicação da regressão linear univariada e da regressão polinomial

```python
model1 = RLA()
model2 = RLGD(learning_rate = 0.009, epochs=1200)
model3 = PR(degree = 2)
model4 = PR(degree = 3)
metrics = Metrics()
```

```python
model1.fit(X_train, y_train)
y_train_predict = model1.predict(X_train)
y_test_predict = model1.predict(X_test)
print("Avaliação: ")
print("Coeficientes: {} e {} ".format(model1.b0, model1.b1))
print("------------------------Treino------------------------")
print("R2: ", metrics.R2(y_true=y_train, y_predict=y_train_predict))
print("MSE: ", metrics.MSE(y_train, y_train_predict))
print("------------------------Teste------------------------")
print("R2: ", metrics.R2(y_test, y_test_predict))
print("MSE: ", metrics.MSE(y_test, y_test_predict))
```

```python
model2.fit(X_train, y_train)
y_train_predict = model2.predict(X_train)
y_test_predict = model2.predict(X_test)
print("Avaliação: ")
print("Coeficientes: {} e {} ".format(model2.b0, model2.b1))
print("------------------------Treino------------------------")
print("R2: ", metrics.R2(y_true=y_train, y_predict=y_train_predict))
print("MSE: ", metrics.MSE(y_train, y_train_predict))
print("------------------------Teste------------------------")
print("R2: ", metrics.R2(y_test, y_test_predict))
print("MSE: ", metrics.MSE(y_test, y_test_predict))
```

```python
model3.fit(X_train, y_train)
y_train_predict = model3.predict(X_train)
y_test_predict = model3.predict(X_test)
print("Avaliação: ")
print("Coeficientes: ", model3.b)
print("------------------------Treino------------------------")
print("R2: ", metrics.R2(y_true=y_train, y_predict=y_train_predict))
print("MSE: ", metrics.MSE(y_train, y_train_predict))
print("------------------------Teste------------------------")
print("R2: ", metrics.R2(y_test, y_test_predict))
print("MSE: ", metrics.MSE(y_test, y_test_predict))
```

```python
model4.fit(X_train, y_train)
y_train_predict = model4.predict(X_train)
y_test_predict = model4.predict(X_test)
print("Avaliação: ")
print("Coeficientes: ",model4.b)
print("------------------------Treino------------------------")
print("R2: ", metrics.R2(y_true=y_train, y_predict=y_train_predict))
print("MSE: ", metrics.MSE(y_train, y_train_predict))
print("------------------------Teste------------------------")
print("R2: ", metrics.R2(y_test, y_test_predict))
print("MSE: ", metrics.MSE(y_test, y_test_predict))
```

O melhor modelo foi a regressão cúbica, isso significa que, provavelmete, a distribuição dos dados possui um formato
mais próximo a uma equação de 3º grau, se ajustando melhor a esse modelo.

## Leitura de trab1_data.csv

```python
data = np.loadtxt("./data/trab1_data.csv", skiprows=1, delimiter=",")
X = data[:,1:-1]
y = data[:, -1]
X = (X - X.min()) / (X.max() - X.min())
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, shuffle=True, test_size=0.2)
```

## Aplicação da regressão linear multivariada

```python
model5 = MLR()
model6 = MLRGD(learning_rate=0.009, epochs=1100, save_epoch_mse=True)
model7 = MLRSGD(learning_rate=0.001, epochs=50, save_epoch_mse=True)
model8 = RMLR(learning_rate=0.1, epochs=1000, _lambda = 1, save_epoch_mse=False)
metrics = Metrics()
```

```python
model5.fit(X_train, y_train)
y_train_predict = model5.predict(X_train)
y_test_predict = model5.predict(X_test)
print("Avaliação: ")
print("Coeficientes: ",model5.b)
print("------------------------Treino------------------------")
print("R2: ", metrics.R2(y_true=y_train, y_predict=y_train_predict))
print("MSE: ", metrics.MSE(y_train, y_train_predict))
print("------------------------Teste------------------------")
print("R2: ", metrics.R2(y_test, y_test_predict))
print("MSE: ", metrics.MSE(y_test, y_test_predict))
```

```python
model6.fit(X_train, y_train)
y_train_predict = model6.predict(X_train)
y_test_predict = model6.predict(X_test)
print("Avaliação: ")
print("Coeficientes: ",model6.b)
print("------------------------Treino------------------------")
print("R2: ", metrics.R2(y_true=y_train, y_predict=y_train_predict))
print("MSE: ", metrics.MSE(y_train, y_train_predict))
print("------------------------Teste------------------------")
print("R2: ", metrics.R2(y_test, y_test_predict))
print("MSE: ", metrics.MSE(y_test, y_test_predict))
```

```python
model7.fit(X_train, y_train)
y_train_predict = model7.predict(X_train)
y_test_predict = model7.predict(X_test)
print("Avaliação: ")
print("Coeficientes: ",model7.b)
print("------------------------Treino------------------------")
print("R2: ", metrics.R2(y_true=y_train, y_predict=y_train_predict))
print("MSE: ", metrics.MSE(y_train, y_train_predict))
print("------------------------Teste------------------------")
print("R2: ", metrics.R2(y_test, y_test_predict))
print("MSE: ", metrics.MSE(y_test, y_test_predict))
```

```python
model8.fit(X_train, y_train)
y_train_predict = model8.predict(X_train)
y_test_predict = model8.predict(X_test)
print("Avaliação: ")
print("Coeficientes: ",model8.b)
print("------------------------Treino------------------------")
print("R2: ", metrics.R2(y_true=y_train, y_predict=y_train_predict))
print("MSE: ", metrics.MSE(y_train, y_train_predict))
print("------------------------Teste------------------------")
print("R2: ", metrics.R2(y_test, y_test_predict))
print("MSE: ", metrics.MSE(y_test, y_test_predict))
```

```python
plt.scatter(range(1,model6.epochs+1), model6.mse, color='black')
plt.plot(range(1,model6.epochs+1), model6.mse, color='blue')

plt.title("Regressão linear usando gradiente descendente")
plt.xlabel("Época")
plt.ylabel("MSE")
plt.show()
```

```python
plt.scatter(range(1,model7.epochs+1), model7.mse, color='black')
plt.plot(range(1,model7.epochs+1), model7.mse, color='blue')

plt.title("Regressão linear usando gradiente descendente estocastico")
plt.xlabel("Época")
plt.ylabel("MSE")
plt.show()
```

O metodo que utiliza gradiente descendente estocástico converge mais rapidamente, necessitando de um menor número de épocas e uma menor taxa de aprendizagem para alcanças valores de rro menores que o gradiente descendente.

```python
mse = np.array([])
_lambda = 6

for l in range(1, _lambda):
    model = RMLR(learning_rate=0.01, epochs=100, _lambda = l, save_epoch_mse=True)
    model.fit(X_train, y_train)
    
    plt.plot(range(1,model.epochs+1), model.mse, color='blue', label='Dados de treino')
    
    model1 = RMLR(learning_rate=0.01, epochs=100, _lambda = l, save_epoch_mse=True)
    model1.fit(X_test, y_test)
    
    plt.plot(range(1,model1.epochs+1), model1.mse, color='green', label='Dados de teste')
    
    plt.title("Regressão linear regularizada com lambda = {}".format(l))
    plt.xlabel("Época")
    plt.ylabel("MSE")
    plt.legend()
    plt.show()
```

O melhor é com lambda igual a 3, onde os valores de erro para treinos e para testes são bem próximos.
