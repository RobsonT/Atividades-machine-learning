# Importação de bibliotecas

```python
import numpy as np
from metrics import Metrics
from models import KNNClassifier, MLP
from sklearn.model_selection import train_test_split
from plots import Plots
```

# Leitura dos dados

```python
data = np.loadtxt("./data1.txt", delimiter=",")
X = data[:, :-1]
y = data[:, -1]
```

# Apresentação dos resultados

```python
def apply_classifier(classifier, clf_name):
    print('________________' + clf_name +'________________')
    Metrics().k_fold(X, y, 5, classifier)
    print('_____________Dados com fronteira de separação_____________')
    Plots().plot_boundaries(X, y, classifier)
```

```python
k_list = range(1, 4)

for k in k_list:
    model1 = KNNClassifier(_k_neighbours=k)
    model1.classes_ = ['X1', 'X2']
    apply_classifier(model1, "KNN com {} vizinhos".format(k))
```

```python
layer_size_list = range(2, 5)

X = (X - X.mean()) / X.std()

for layer_size in layer_size_list:
    model2 = MLP(_epochs=300, _learning_rate=0.0001, _layer_size=layer_size)
    model2.classes_ = ['X1', 'X2']
    apply_classifier(model2, "MLP com {} unidades".format(layer_size))
```
